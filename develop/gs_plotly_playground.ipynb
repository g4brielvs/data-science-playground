```python
from plotly.graph_objs import *
# Plotting with Pandas
import pandas as pd
```

```python
df = pd.read_csv('https://plot.ly/~etpinard/191.csv')
df.head()
```

```python
continents = ['Africa', 'Americas', 'Asia', 'Europe', 'Oceania']
attributes = ['x', 'y', 'text', 'size']

tags = ['Gross Domestic Product per Capita [in USD of the year 2000]', 
        'Life Expentancy [in years]', 
        'text', 
        'size']

ts = dict(zip(attributes, tags))

list_ts = list()
for k, v in ts.items():
    l = [(f'{c}, {k}', f'{c}_{ts[k]}') for c in continents]
    list_ts.extend(l)

d = dict(list_ts)
```

```python
df = df.rename(index=str, columns=d)

for c in continents:
    df[f'{c}_size'] = df[f'{c}_size'].fillna(0.0).astype(int)
```

```python
iplot({
       'data': [
        Scatter(x=df[continent+'_Life Expentancy [in years]'],
                y=df[continent+'_Gross Domestic Product per Capita [in USD of the year 2000]'],
                text=df[continent+'_text'],
                marker=Marker(size=df[continent+'_size'], sizemode='area', sizeref=131868,),
                mode='markers',
                name=continent) for continent in continents
        ],
        'layout': Layout(xaxis=XAxis(title='Life Expectancy'), yaxis=YAxis(title='GDP per Capita', type='log'))
        
    }, show_link=False)
```
